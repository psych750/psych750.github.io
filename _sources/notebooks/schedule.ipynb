The markdown file should contain sufficient narrative to explain your analyses and graphs. **Please do not** just have a sequence of R code blocks. Include text above/below explaining what's going on, sequencing it in a logical way so that a reader can proceed from the beginning and understand what each analysis and graph that follows is conveying. **Please do** organize the markdown file neatly using nested headings (#, ## etc.).
